```python
import requests

# URL de la imatge
url = 'http://consultas2.oepm.es/ceo/ImagenMarcaServlet?urlMarca=http://imagensignos.oepm.local/imagenes/000183/0000719581.JPG'

# Nom del fitxer de destinació
file_name = 'imatge_prova.jpg'

# Realitza la sol·licitud HTTP GET per descarregar la imatge
response = requests.get(url)

# Comprova que la sol·licitud va ser exitosa
if response.status_code == 200:
    # Escriu el contingut de la imatge en un fitxer
    with open(file_name, 'wb') as file:
        file.write(response.content)
    print(f"La imatge ha estat descarregada i guardada com {file_name}.")
else:
    print("No s'ha pogut descarregar la imatge. Estat:", response.status_code)
```

```python
import requests
import random
import time

for index, row in df2.iterrows():

  url = row['Imatge'].strip()
  codi = str(row['Codi'])

  # Nom del fitxer de destinació
  file_name = "M"+codi+"."+url.split('.')[-1]

  # Realitza la sol·licitud HTTP GET per descarregar la imatge
  response = requests.get(url)

  # Comprova que la sol·licitud va ser exitosa
  if response.status_code == 200:
      # Escriu el contingut de la imatge en un fitxer
      with open(file_name, 'wb') as file:
          file.write(response.content)
      print(f"La imatge de la marca {row['Denominació']} ha estat descarregada i guardada com {file_name}.")
  else:
      print(f"No s'ha pogut descarregar la imatge de la marca {row['Denominació']}. Estat:", response.status_code)

  time.sleep(random.uniform(5, 10))
```

```python
import zipfile
import io
from google.colab import files
import os

# Ruta de la carpeta on es troben les imatges
folder_path = '/content'

# Crear un fitxer ZIP i afegir la imatge
zip_filename = 'OEPM_Imatges_NO_GPG.zip'

# Crear el fitxer ZIP
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    # Iterar sobre tots els arxius de la carpeta
    for root, dirs, files_list in os.walk(folder_path):
        for file in files_list:
            # Comprova si l'arxiu és una imatge (per extensió)
            if file.lower().strip().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tif')):
                file_path = os.path.join(root, file)
                # Afegir l'arxiu al fitxer ZIP
                zipf.write(file_path, os.path.relpath(file_path, folder_path))

# Descarregar el fitxer ZIP
files.download(zip_filename)
```
